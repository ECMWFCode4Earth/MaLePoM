#07_2019

We upload the weather and CO dataset for july 2019 and we prepare it for the model

```python
dataset_07_19 = xr.open_dataset('Data/07_2019_weather_and_CO.nc')
dataset_07_19 = dataset_07_19.sortby(['latitude','longitude','time'])  
dataframe_07_19 = dataset_07_19.to_dataframe()
dataframe_07_19 = dataframe_07_19.reset_index()
dataframe_07_19 = dataframe_07_19.set_index('time')
```

Here we created the shifted features

```python
dataframe_07_19['tmp_shift_1']=dataframe_07_19.groupby(['latitude','longitude']).tmp.shift(periods=1)
dataframe_07_19['tot_wind_shift_12']=dataframe_07_19.groupby(['latitude','longitude']).tot_wind.shift(periods=12)
dataframe_07_19['rel_hum_shift_8']=dataframe_07_19.groupby(['latitude','longitude']).rel_hum.shift(periods=8)

dataframe_07_19 = dataframe_07_19.reset_index()
dataframe_07_19 = dataframe_07_19.set_index(['latitude','longitude','time'])
```

```python
ds_final_07_19=dataframe_07_19.to_xarray()
ds_final_07_19
```

Now we save the dataset

```python
ds_final_07_19.to_netcdf('Data/07_2019_weather_and_CO_for_model.nc', 'w', 'NETCDF4')
```
