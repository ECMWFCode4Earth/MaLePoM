#Load the datasets

I apply the following changes to all three datasets that I upload:
*   cut_region function: the dataset contains a larger region than the one we are analyzing, so we proceed to eliminate the excess area
*   rescale function: we change the scale of the dataset from 0.75 degree to 0.25 degree
*   resample function: we create a time series with an hourly frequency

Extremes of the region that we are analyzing:

```python
lat_s = 43.0
lat_n = 51.0
lon_e =  4.0
lon_w = 12.0
```

##Wind_vegetation_pressure_cloud

```python
wvpc_instance = CustomDataset('Data/05_2019_wind_vegetation_pressure_cloud.nc')
wvpc_instance.cut_region(lat_n, lat_s, lon_e, lon_w)
wvpc_instance.rescale()
wvpc_instance.resample("1H")
wvpc_Dataset = wvpc_instance.get_dataset()
wvpc_Dataset
```

##Hum_temp

```python
hum_temp_instance = CustomDataset('Data/05_2019_hum_temp.nc')
hum_temp_instance.cut_region(lat_n, lat_s, lon_e, lon_w)
hum_temp_instance.rescale()
hum_temp_instance.resample("1H")
hum_temp_instance.get_dataset()
```

Rename the variables

```python
hum_temp_instance.rename_var({'t':'tmp',
                              'q':'sp_hum',
                              'r':'rel_hum'})
hum_temp_Dataset = hum_temp_instance.get_dataset()
hum_temp_Dataset
```

##Water

```python
water_instance = CustomDataset('Data/05_2019_water.nc')
water_instance.cut_region(lat_n, lat_s, lon_e, lon_w)
water_instance.rescale()
water_instance.resample("1H")
water_Dataset = water_instance.get_dataset()
water_Dataset
```

#Put the datasets togheter

```python
weather_dataset = wvpc_Dataset.merge(hum_temp_Dataset)
weather_dataset = weather_dataset.merge(water_Dataset)
weather_dataset
```

#New feature

I want to add a new feature: is the sum of the two direction wind

```python
weather_dataset['tot_wind']= np.sqrt(np.square(weather_dataset['u10']) + np.square(weather_dataset['v10']))
weather_dataset
```

#Save dataset

```python
weather_dataset.to_netcdf('Data/05_2019_weather.nc', 'w', 'NETCDF4')
```
